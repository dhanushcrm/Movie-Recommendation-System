Importing necessary libraries and dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Pre-preprocessing

```python
movies_data = pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
#Selecting relevant features (Feature selection)

selected_features = ['genres','keywords','tagline','cast','director']
```

```python
#Replacing null values
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
combined_features = movies_data['genres']+''+movies_data['keywords']+''+movies_data['tagline']+''+movies_data['cast']+''+movies_data['director']
```

```python
print(combined_features)
```

```python
#Converting text to numerical (feature vector)

vectorize = TfidfVectorizer()
```

```python
feature_vectors = vectorize.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity

```python
#Getting Similarity Score using cosine similarity

similarity = cosine_similarity(feature_vectors)
```
